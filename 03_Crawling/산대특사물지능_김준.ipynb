### 3번 문제

```python
from selenium.webdriver.chrome.service import Service 
from webdriver_manager.chrome import ChromeDriverManager 
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
import time as tm
import pandas as pd
from tqdm import tqdm as tq 


s=Service(ChromeDriverManager().install())
driver=wb.Chrome(service=s)
driver.get('https://www.gmarket.co.kr/n/best')

lst1=[]
lst2=[]


for i in tq(range(20)):
    image=driver.find_elements(By.CLASS_NAME,'image__lazy.image__load')
    image[i].click()
    
    tm.sleep(1)
    title = driver.find_element(By.CSS_SELECTOR, "#itemcase_basic > div.box__item-title > h1").text
    price = driver.find_element(By.CSS_SELECTOR, "#itemcase_basic > div.box__item-title > div.price > span:nth-child(2) > strong").text
    
    lst1.append(title)
    lst2.append(price)
    
    tm.sleep(1)
    driver.back()

driver.quit()

df=pd.DataFrame({'상품명':lst1,'가격':lst2})
df
```

### 4번 문제

```python
import requests as req
from selenium.webdriver.chrome.service import Service 
from webdriver_manager.chrome import ChromeDriverManager 
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
import time as tm
import pandas as pd
from tqdm import tqdm as tq 
h={'User-Agent':
'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36'}
a=req.get('https://www.melon.com/chart/',headers=h)

b=bs(a.text,'lxml')

b.select('div.ellipsis.rank01')
b.select('div.ellipsis.rank02')

# b.select('span.cnt>span')
song_name=[]
singer=[]
rank=[i for i in range(1,101)]
# good=[] 좋아요의 갯수가 담겼을 리스트

for i in range(len(b.select('div.ellipsis.rank02'))):
    song_name.append(b.select('div.ellipsis.rank01')[i].text.strip())
    singer.append(b.select('span.checkEllipsis')[i].text.strip())
#    lst4.append(b.select('button.button_etc.like>span.cnt')[i].text.strip())
# 좋아요가 될뻔했던 아이

df = pd.DataFrame({"song":song_name, "singer":singer,"순위":rank})
df.set_index('순위',inplace=True)
df.to_csv("멜론 100.csv",encoding='utf-8')
# jupyter내부에서 열어볼 경우 encoding='utf-8' 
# 외부 프로그램에서 열어볼 경우 encoding='euc-kr'
```

### 5번 문제

```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time as tm
import pandas as pd
```

```python
s= Service(ChromeDriverManager().install())
driver=wb.Chrome(service=s)

keyword=input('유튜브 검색어를 입력하세요>>>')
url= f'https://www.youtube.com/results?search_query={keyword}'
driver.get(url)

video_url=driver.find_elements(By.CSS_SELECTOR,"a#video-title")
for i in video_url:
    print(i.get_attribute('href'))

names=driver.find_elements(By.CSS_SELECTOR,"a#video-title")
for i in names:
    print(i.text)

y_title=[]
y_link=[]
for i in names:
    y_title.append(i.text)
for u in video_url:
    y_link.append(u.get_attribute('href'))

y_df=pd.DataFrame({"제목":y_title,"링크":y_link})
y_df.to_csv('유튜브 제목,링크.csv',encoding='utf-8' )
# jupyter내부에서 열어볼 경우 encoding='utf-8' 
# 외부 프로그램에서 열어볼 경우 encoding='euc-kr'

driver.quit()
```
