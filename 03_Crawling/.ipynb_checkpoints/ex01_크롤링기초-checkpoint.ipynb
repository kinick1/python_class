## requests 라이브러리 사용법

```python
# 라이브러리 불러오기
import requests as req
# 서버에게 데이터를 요청 및 응답받기 위한 라이브러리
```

```python
res=req.get('https://www.naver.com/')
res
# Response [200]번대 : 성공적으로 요청 및 응답이 이루어졌다는 의미
# Response [400]번대 : 클라이언트가 요청에서 오류가 발생했을 때 
# Response [500]번대 : 서버가 응답 오류가 발생했을 때
```

```python
res.text
# html 문서가 문자열 형탤 가져와졌다
```

```python
res=req.get('https://www.melon.com/')
res
# 406 에러 : 클라이언트부분에서 오류 발생 -> 멜론 페이지에서 파이썬 코드로 접근하는 것을 막고 있기 때문에
# 사람이 접속하는것처럼 속이는 우회접속
```

```python
h={"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36'}
```

```python
req.get('https://www.melon.com/',headers=h)

# User-Agent를 이용한 우회접속 방법
# 페이지 불러오기 - >코드에 오류가 있지도 않은데 400번대 오류가 발생 시
# F12 버튼(개발자 도구)Network 탭으로 접근
# F5 눌러서 새로고침한 후 Document 타입의 html 문서정보로 접근
# html 문서정보 클릭후 User-Agent에 대한 정보를 복사해서
# 파이썬 코드에 딕셔너리형태로 저장후 넣어준다
# 페이지를 불러올때  headers 값에 넣어주면 끝
```

### beautifulSoup 라이브러리

```python
from bs4 import BeautifulSoup as bs
```

```python
res= req.get('https://www.kakaocorp.com/page/service/service/KakaoTalk')
res
```

```python
res.text
```

```python
# bs(parsing할 html 문서, 'parsing할 방법')
# 파이썬에서 제공하는 parsing 방법은 총 4가지
# lmxl : 빠르고 관대하기때문

soup=bs(res.text,'lxml') # 파이썬 객체 형태로 변환
```

### 카카오톡 페이지 정보를 가져와보자
- 카카오톡 페이지에서 글자를 가져오자

```python
# 카톡 안녕가이드 라는 글자를 수집
#  태그의 위치를 찾는 방법
# 1. 내가 가져오고 싶은 contents 위에 마우스를 올려두고 우클릭 -> 검사
# 2. 개발자도구 상단에 커서 모양 아이콘 클릭후 가져오고싶은 콘텐츠 클릭
```

```python
# 카독 안녕가이드는 a태그 안에 있다
# a태그를 가져오는 방법

soup.select('a')
```

```python
# a태그 중에서 우리가 원하는 태그만 가져오려면 선택자가 필요
# a태그 중에서 클래스가 link_direct인 얘들만 데러오가
# 클래스에 접근하는 기호는 온점(.)을 사용
```

```python
soup.select('a.link_direct') # ->데이터형식 : list
# select : 여러가지 요소를 복수형으로 가져와줌 (항상 리스트형태)
```

```python
# select_one : 여러가지 요소중에서 맨위에 있는 첫번째 요소만 단수형으로 가져와준다
soup.select_one('a.link_direct')
```

```python
# 카톡 안녕가이드 글자만 수집
a=soup.select('a.link_direct')[0]
```

```python
print(a.text) # .text : 요소 내부에 있는 콘텐츠만 가죠오는 기능
# 데이터 앞 뒤로 공백이 많다.
```

```python
print(a.text.strip())
```
