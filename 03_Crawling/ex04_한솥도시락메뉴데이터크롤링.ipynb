#### 실습! 한셀에 완벽한 크롤링 코드 짜기!
1. 한솥 메뉴 사이트 요청
2. 더보기 끝까지 누르기
3. 가격과 메뉴명 요소 수집
4. content 추출
5. DataFrame 제작 & 형식 저장
6. 크롬 브라우저 닫기
7. 한셀로 합치기 (shift + m)

````python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
import time
import pandas as pd


driver=wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')


while(1):
    try:
        btn=driver.find_element(By.CLASS_NAME,'c_05')
        btn.click()
        time.sleep(1)
    except:
        print('더보기 없음')
        break

driver.page_source

a=bs(driver.page_source,'lxml')
a

menu=a.select('h4.h.fz_03')
len(menu)

price=a.select('div.item-text > div > strong')
price_c=len(price)
price_c

lst1=[]
lst2=[]

for i in range(price_c):
    lst1.append(menu[i].text)
    lst2.append(price[i].text)
lst1,lst2

df = pd.DataFrame({"메뉴":lst1, "가격":lst2})
df

df.to_csv('한솥메뉴.csv',encoding='utf-8')
````
