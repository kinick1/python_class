```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

from selenium.webdriver.support.ui import WebDriverWait
# WebDriverWatit : 특정 조건이 충족될때까지 기다리는데 사용하는 모듈
from selenium.webdriver.support import expected_conditions as EC
# EC : WebDriver가 기다리는 조건을 정의 하는데 사용하는 모듈

import time as tm
import pandas as pd
```

```python
s= Service(ChromeDriverManager().install())
driver=wb.Chrome(service=s)

keyword=input('유튜브 검색어를 입력하세요>>>')
url= f'https://www.youtube.com/results?search_query={keyword}'
driver.get(url)
```

```python
names=driver.find_elements(By.CSS_SELECTOR,'yt-formatted-string.style-scope.ytd-video-renderer')
for i in names:
    print(i.text)
    
# 유튜브 제목만 출력된게 아니다.
# 페이지를 잘 살펴보면 class 값을 여러 태그들이 지니고 있다.
#  -> 다른 방법을 사용
```

```python
# a태그에 감싸져 있다 -> a태그와 클래스까지 활용
```

```python
# url 수집
# 같은 id를 가진 a태그가 여러개 존재한다? -->id가 class처럼 사용된다
video_url=driver.find_elements(By.CSS_SELECTOR,"a#video-title")
# 유튜브에 있는 영상 하나하나가 전부 다른 서버에 존재하기 때문
for i in video_url:
    print(i.get_attribute('href'))
```

```python
len(video_url), len(names)
# names와 video의 길이가 다르다
# 수집 방법의 차이가 존재한다
```

```python
names=driver.find_elements(By.CSS_SELECTOR,"a#video-title>yt-formatted-string")
for i in names:
    print(i.text)
```

```python
names=driver.find_elements(By.CSS_SELECTOR,"a#video-title")
for i in names:
    print(i.text)
```

```python
lst1=[]
lst2=[]

for i in names:
    lst1.append(i.text)


for u in video_url:
    lst2.append(u.get_attribute('href'))
```

```python
pd.set_option('display.max_colwidth',None)
```

```python
dic={"영상이름":lst1,"링크":lst2}
df=pd.DataFrame(dic)
df
```

```python
df.to_csv("유튜브 데이터.csv")
```
