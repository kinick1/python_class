## matplotlib
- 2D 그래프로 시각화가 가능한 라이브러리
- 파이썬에서 데이터를 차트나 플롯으로 그려줌
- 차트의 종류는 산점도, 바, 파이, 히스토그램, 꺽은선 그래프 등으로 다양하게 존재

### 데이터 시각화
- 광범위하게 분산된 방대한 양의 자료를 하눈에 볼 수 있도록 도표나 차트등으로 정리하는 것
- 시각화를 통해 데이터의 특징을 쉽게 파악할수 있다.

```python
# matplotlib 패키지 불러오기 (패키지를 불러온 후 pyplot이라는 라이브러리를 사용하겠다.)
# 이름은 plt로 부르겠다
import matplotlib.pyplot as plt
```

### 기본적인 그래프 그리기

```python
# Line Plot(선 그래프)
y=[2,12,6,7]
plt.plot(y)

# plot(좌표의 y요소)
# x좌표는 0부터 1씩 증가하며 자동으로 할당

plt.show()
# plt.show() ->그동안 그려왔던 plot을 출력하겠다.
# print와 같은 기능을 한다. -> 그래프를 그릴 때는 항상 마지막엔 show로 마무리를 지어야함
```

```python
# x축과 y축을 전부 설정해서 그려보자
# 선 스타일도 입혀보자
x=[1,2,3]
y=[3,4,5]
plt.plot(x,y,ls='--')
plt.show()
```

```python
# x축 y축 설정해서 그리고 marker 찍기
x=[1,2,3]
y=[3,4,5]
plt.plot(x,y,ls='--',marker='p')
plt.show()
```

```python
# x축 y축 설정해서 선 두께 설정
x=[1,2,3]
y=[3,4,5]
plt.plot(x,y,ls='--',marker='|',lw=1)
plt.show()
```

```python
# x축 y축 설정해서 선 색깔 설정
x=[1,2,3]
y=[3,4,5]
plt.plot(x,y,ls='--',marker='p', color='c')
plt.show()

# 속성값
# 1. 라인의 스타일 설정 : ls(line style )-> 문자열
# 2. 라인의 두께 설정 : lw(line width)-> 숫자
# 3. 선의 색 바꾸기 : c(color)-> 선 색과 마커의 색까지 모두 변경
# 4. 마커 내부 색 바꾸기 : mfc(maker face color) -> 문자열
# 5. 마커 선 색깔 : mec( marker edge color) -> 문자열
# 6. 마커 크기 : ms ->숫자
# 7. 마커의 선 굴기 : mew( marker edge width)->숫자
```

### 선 그래프 실습
- x=np.arange(8)
- y=[5,2,3,6,8,7,5,7]

```python
import numpy as np
```

```python
x=np.arange(8)
y=[5,2,3,6,8,7,5,7]

plt.plot(x,y,ls='-.',c='c',marker='$★$',mec='skyblue',mfc='w',lw=4,ms=20)
plt.show()
```

```python
x=np.arange(8)
y=[5,2,3,6,8,7,5,7]

plt.plot(x,y,ls='-.',c='#FF3366',marker='$★$',mec='skyblue',mfc='w',lw=4,ms=20)

# 그래프의 숫자 표시 범위 지정
# 범위를 지정하여 그래프를 축소하거나 확대할수 있다.

plt.xlim(-2.10)
plt.ylim(-5,12)

plt.show()
```

```python
x=np.arange(8)
y=[5,2,3,6,8,7,5,7]

plt.plot(x,y,ls='-.',c='#FF3366',marker='$★$',mec='skyblue',mfc='w',lw=4,ms=20)

# 그리드 그리기 (그리드 : 모눈종이, 격자눈금) -> 그래프를 보기 쉽게 하기 위해
plt.grid()

# 눈금 출력 방법 -> ticks
plt.xticks(x) # ->y데이터를 기준으로 눈금을 그려라

plt.show()
```

### 여러개 차트를 하나의 공간에 그리기

```python
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]

plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')

# 범례 설정하기
# 범례 : 데이터의 종류를 표시하기 위한 텍스트
plt.legend()

# 차트 제목 설정하기
# loc : 위치 설정(left, right, center)
plt.title('촤트',loc='right')

# x축과 y축에 이름을 넣어보자
plt.xlabel('x')
plt.ylabel('y')


plt.show()
```

### bar 그래프 그리기

```python
# bar
x=[0,1,2,3,4,5,6]
y=[80,90,70,50,60,90,100]

plt.bar(x,y)

plt.xlabel('강의장')
plt.ylabel('점수')

plt.show()
```

```python
# barh 그래프

x=[0,1,2,3,4,5,6]
y=[80,90,70,50,60,90,100]

plt.barh(x,y)

plt.xlabel('점수')
plt.ylabel('강의장')

plt.show()
```

```python
# bar 그래프
x=[0,1,2,3,4,5,6]
y=[80,90,70,50,60,90,100]

plt.bar(x,y)

plt.xticks(x,['1강의장','2강의장','3강의장','4강의장','5강의장','6강의장','7강의장'])
# ylim를 설정해서 분포된 데이터 확대
plt.ylim(40,100)

plt.xlabel('강의장')
plt.ylabel('점수')

plt.title('CGI 강의장별 시험 성적',loc='right')

plt.show()
```

### 공공데이터 분석 및 시각화
- 2017년 전국 교통사고 데이터

```python
# 데이터 분석에 필요한 라이브러리 로딩

# pandas, numpy, maplotlib.pyplot
import pandas as pd # 수치 해석 전용 라이브러리
import numpy as np # 데이터를 표 형태로 제공하는 라이브러리
import matplotlib.pyplot as plt # 데이터를 시각화를 도와주는 라이브러리
```

```python
# 데이터 로딩
data=pd.read_csv('Traffic_Accident_2017.csv',encoding='euc-kr')
data
```

```python
# 데이터의 규모 확인
# info(): 데이터 프레임에서 컬럼이름 NaN값이 아닌 갯수, 데이터 타입
data.info()
```

```python
# 데이터 크기 확인
data.shape
```

```python
# 데이터프레임 최상단 5개만 출력하는 방법
# 기본값=5
# 원하는 만큼 숫자를 입력해 출력 조절 가능
data.head(3)
```

```python
# 데이터 프레임 최하단 5개만 출력하는 방법
# 기본값=5
# 원하는 만큼 숫자를 입력해 출력 조절 가능
data.tail()
```

```python
# 신고 건수로 분류 하여 갯수 새기
# value_counts : 하나의 컬럼 안에 있는 유니크 값들과 그 값들의 개수를 각각 출력
data['부상신고자수'].value_counts()
```

### 요일별 사고건수 카운트
- 요일 순서대로 정렬후 변수에 저장(y)

```python
# 요일별 발생한 사고 건수를 카운트 해보자
temp=data['요일'].value_counts()
temp
```

```python
# 월요일~ 금요일까지 순서대로 정렬
temp.sort_index() #정렬을 하더라도 자음순서로 정렬 됨
```

```python
# 유니크 값으로 설정 후 변수에 담아주기
y1=temp[['월','화','수','목','금','토','일']]
y1
```

### 시각화
- 바 그래프

```python
# y축= y -> 500~620
# x축= 7
# 그래프 제목 :  전체 요일별 사망교통사고 건수
# x축 : 요일
# y축 : 사고건수
```

```python
x=['월','화','수','목','금','토','일']
y=y1.values

plt.bar(x,y)
plt.ylim(500,620)
plt.yticks(y)
plt.title('전체 요일별 사망교통사고 건수',loc='right')
plt.xlabel('요일')
plt.ylabel('사고건수            ',rotation=0)
plt.show()
```

```python
days = np.arange(7)
counts = y.values


plt.bar(days, counts)
plt.xticks(days,['월', '화', '수', '목', '금', '토', '일'])
plt.yticks(counts) # y축의 값이 어떤값인지 정확하고 직관적으로 알려주기 위함
plt.ylim(500,620) # y축 범위 :500~620
plt.title('전체 요일별 사망교통사고 건수',loc='right')
plt.xlabel('요일')
plt.ylabel('사고건수             ',rotation=0) # 사고건수라는 글자를90도로 세워보자 
# rotation은 회전 이라는 뜻으로 y라벨에 해당하는 글자를 얼마나 회전할 것인지 기입(숫자)
 
plt.show()
```
