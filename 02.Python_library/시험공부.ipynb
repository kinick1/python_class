```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data2017=pd.read_csv('Traffic_Accident_2017.csv',encoding='euc-kr')
```

```python
data2017
```

```python
data2017.shape
```

```python
data2017['경상자수'].value_counts()
```

```python
temp=data2017['요일'].value_counts()
y1=temp[['월','화','수','목','금','토','일']]
y1
```

```python
x=['월','화','수','목','금','토','일']
y=y1.values

plt.bar(x,y)
plt.ylim(500,620)
plt.yticks(y)
plt.title('전체 요일별 사망교통사고 건수',loc='right')
plt.xlabel('요일')
plt.ylabel('사고건수            ',rotation=0)
plt.show()
```
